```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from transformers import AutoImageProcessor, AutoModel
from PIL import Image
import pandas as pd
from sklearn.metrics import accuracy_score
from tqdm import tqdm
```

```python

# === Define your Multi-Head Model ===
class MultiHeadViT(nn.Module):
    def __init__(self, model_name, num_labels_list):
        super(MultiHeadViT, self).__init__()
        self.backbone = AutoModel.from_pretrained(model_name)
        hidden_size = self.backbone.config.hidden_size
        self.heads = nn.ModuleList([
            nn.Linear(hidden_size, num_labels) for num_labels in num_labels_list
        ])

    def forward(self, pixel_values):
        outputs = self.backbone(pixel_values=pixel_values)
        cls_token = outputs.last_hidden_state[:, 0, :]
        logits = [head(cls_token) for head in self.heads]
        return logits

# === Dummy Dataset ===
class DummyDataset(Dataset):
    def __init__(self, image_paths, labels_list, processor):
        self.image_paths = image_paths
        self.labels_list = labels_list
        self.processor = processor

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        image = Image.open(self.image_paths[idx]).convert("RGB")
        inputs = self.processor(images=image, return_tensors="pt")

        labels = [torch.tensor(x) for x in self.labels_list[idx]]  # FIXED

        return {
            'pixel_values': inputs['pixel_values'].squeeze(0),
            'labels': labels
        }


# === Instantiate everything ===
vis_model_name = "google/vit-base-patch16-224"
num_labels_list = [11, 2, 3, 8, 2]  # Task 1: 10 classes, Task 2: 5 classes, Task 3: 3 classes

vis_model = MultiHeadViT(vis_model_name, num_labels_list)

processor = AutoImageProcessor.from_pretrained(vis_model_name)

# Dummy paths and labels

# df = pd.read_csv("/content/drive/MyDrive/Skin Disease Dataset/training split/temp train.csv")
# df = pd.read_csv("/content/drive/MyDrive/Thesis/Skin Disease Dataset/training split/temp train.csv")
df = pd.read_csv("/content/drive/MyDrive/Thesis/Skin Disease Dataset/training split/Train.csv")

img_path_lst = []
labels_list = []


for idx, row in df.iterrows():
    # image_path = img_path_lst.append("/content/drive/MyDrive/Skin Disease Dataset/training split/combined-img/" + row['image'])
    image_path = img_path_lst.append("/content/drive/MyDrive/Thesis/Skin Disease Dataset/training split/combined-img/" + row['image'])
    labels = labels_list.append([int(row['name']), int(row['cancerous']), int(row['severity']), int(row['cause']), int(row['contagious'])])

dataset = DummyDataset(img_path_lst, labels_list, processor)
dataloader = DataLoader(dataset, batch_size=2, shuffle=True)

# Loss and optimizer
criterions = [nn.CrossEntropyLoss() for _ in num_labels_list]  # One loss per head
optimizer = optim.Adam(vis_model.parameters(), lr=2e-5)

# === Training loop ===
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
vis_model.to(device)

vis_model.train()

for epoch in tqdm(range(10), desc="Epochs"):  # 3 epochs for example
    total_loss = 0

     # Initialize accuracy counters
    correct_preds = [0] * len(num_labels_list)
    total_preds = [0] * len(num_labels_list)

    for batch in dataloader:
        pixel_values = batch['pixel_values'].to(device)
        labels = [label.to(device) for label in batch['labels']]

        optimizer.zero_grad()

        logits_list = vis_model(pixel_values)

        # Compute individual losses
        losses = []
        for i, (logits, label, criterion) in enumerate(zip(logits_list, labels, criterions)):
            loss = criterion(logits, label)
            losses.append(loss)

            # === Accuracy calculation ===
            preds = torch.argmax(logits, dim=1)
            correct_preds[i] += (preds == label).sum().item()
            total_preds[i] += label.size(0)

        # Total loss = sum of all task losses
        loss = sum(losses)

        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch {epoch+1} - Loss: {total_loss:.4f}")
    for i in range(len(num_labels_list)):
        acc = correct_preds[i] / total_preds[i] if total_preds[i] > 0 else 0
        print(f"  Task {i+1} Accuracy: {acc * 100:.2f}%")
```

```python
########### Grad-CAM test

# Storage
activations = None
gradients = None

def save_activation(module, input, output):
    # output is a tuple: (last_hidden_state,)
    global activations
    activations = output[0]    # shape = (1, 197, hidden)

def save_gradient(module, grad_input, grad_output):
    global gradients
    gradients = grad_output[0]  # gradient wrt last_hidden_state



# # REGISTER CORRECT LAYER (IMPORTANT!)
# target_layer = vis_model.backbone.encoder.layer[-1]
# target_layer.register_forward_hook(save_activation)
# target_layer.register_backward_hook(save_gradient)


# target_layer = vis_model.backbone.encoder.layer[-1].output
# target_layer.register_forward_hook(save_activation)
# target_layer.register_backward_hook(save_gradient)



target_layer = vis_model.backbone.encoder.layer[-1].attention.output
target_layer.register_forward_hook(save_activation)
target_layer.register_full_backward_hook(save_gradient)






import numpy as np
import cv2
from PIL import Image

def generate_gradcam(img_path, head_id=0):

    vis_model.eval()

    # Load image
    image = Image.open(img_path).convert("RGB")
    inputs = processor(images=image, return_tensors="pt").to(device)

    # Forward pass
    logits_list = vis_model(inputs["pixel_values"])
    logits = logits_list[head_id]   # choose which head (0–4)

    pred_class = logits.argmax()

    # Backward: compute gradient of predicted class
    vis_model.zero_grad()
    logits[0, pred_class].backward(retain_graph=True)

    global activations, gradients

    # activations: (1, 197, hidden)
    # gradients:   (1, 197, hidden)

    # === Compute CAM ===
    weights = gradients.mean(dim=-1).squeeze()  # (197,)
    cam = (weights.unsqueeze(-1) * activations.squeeze()).sum(-1)  # (197,)

    cam = cam[1:]  # remove CLS token → keep only 196 patches
    cam = cam.reshape(14, 14).detach().cpu().numpy()

    cam = np.maximum(cam, 0)
    cam = cam / cam.max()

    # Resize to image size
    cam = cv2.resize(cam, (image.width, image.height))

    # Color map
    heatmap = cv2.applyColorMap(np.uint8(255 * cam), cv2.COLORMAP_JET)
    heatmap = cv2.cvtColor(heatmap, cv2.COLOR_BGR2RGB)

    # Blend
    img_np = np.array(image)
    superimposed = cv2.addWeighted(img_np, 0.6, heatmap, 0.4, 0)

    return superimposed, cam, pred_class.item()




img_path = '/content/drive/MyDrive/Thesis/Skin Disease Dataset/training split/combined-img/ISIC_7623407.jpg'

cam_img, raw_cam, pred_cls = generate_gradcam(img_path, head_id=0)

print("Predicted class:", pred_cls)

import matplotlib.pyplot as plt
plt.figure(figsize=(8,8))
plt.imshow(cam_img)
plt.axis('off')
plt.show()
```

```python
# import torch
# import torch.nn.functional as F
# import numpy as np
# import cv2
# from PIL import Image
# import matplotlib.pyplot as plt

# # ------------------------------------------
# # ATTENTION ROLLOUT FOR YOUR ViT ARCHITECTURE
# # ------------------------------------------

# attn_maps = []   # store attention from hooks

# def save_attn_hook(module, input, output):
#     """
#     output: (batch, num_heads, tokens, tokens)
#     """
#     attn_maps.append(output.detach().cpu())

# def register_attn_hooks(model):
#     hooks = []
#     for layer in model.backbone.encoder.layer:
#         attn = layer.attention.attention   # ViTSelfAttention inside ViTAttention
#         hooks.append(attn.register_forward_hook(save_attn_hook))
#     return hooks


# def attention_rollout(vis_model, processor, img_path, head_fusion="mean", discard_ratio=0.0):

#     # Clear any existing residuals
#     global attn_maps
#     attn_maps = []

#     # Register hooks
#     hooks = register_attn_hooks(vis_model)

#     # Preprocess image
#     img = Image.open(img_path).convert("RGB")
#     inputs = processor(images=img, return_tensors="pt")

#     # Forward pass — hooks will capture attention maps
#     with torch.no_grad():
#         _ = vis_model(**inputs)

#     # Remove hooks
#     for h in hooks:
#         h.remove()

#     # Convert attention maps
#     attn = torch.stack(attn_maps)
#     # shape: [num_layers, batch, heads, tokens, tokens]
#     num_layers, _, num_heads, t1, t2 = attn.shape

#     # Head fusion
#     if head_fusion == "mean":
#         attn = attn.mean(dim=2)  # avg over heads
#     elif head_fusion == "max":
#         attn, _ = attn.max(dim=2)
#     else:
#         raise ValueError("head_fusion must be 'mean' or 'max'")

#     # Attention rollout
#     eye = torch.eye(attn.size(-1))
#     attn = attn + eye  # add residual connections
#     attn = attn / attn.sum(dim=-1, keepdim=True)

#     joint_attn = attn[0]
#     for i in range(1, num_layers):
#         joint_attn = attn[i].bmm(joint_attn)  # multiply attention matrices

#     # Token → patch attention from CLS token
#     cls_attn = joint_attn[0, 0, 1:]  # remove CLS itself

#     num_patches = int(np.sqrt(cls_attn.shape[0]))
#     mask = cls_attn.reshape(num_patches, num_patches).numpy()

#     # Resize heatmap to image size
#     heatmap = cv2.resize(mask, img.size)
#     heatmap = heatmap - heatmap.min()
#     heatmap = heatmap / (heatmap.max() + 1e-8)

#     # Overlay heatmap
#     heatmap_color = cv2.applyColorMap((heatmap * 255).astype(np.uint8), cv2.COLORMAP_JET)
#     heatmap_color = cv2.cvtColor(heatmap_color, cv2.COLOR_BGR2RGB)

#     blended = (0.5 * np.array(img) + 0.5 * heatmap_color).astype(np.uint8)

#     return blended, heatmap


def attention_rollout(vis_model, processor, img_path, head_fusion="mean", discard_ratio=0.0):

    global attn_maps
    attn_maps = []

    hooks = register_attn_hooks(vis_model)

    # Load image
    img = Image.open(img_path).convert("RGB")
    inputs = processor(images=img, return_tensors="pt")

    # 🔥 FIX: Move input tensors to same device as model
    device = next(vis_model.parameters()).device
    inputs = {k: v.to(device) for k, v in inputs.items()}

    # Forward pass triggers hook
    with torch.no_grad():
        _ = vis_model(**inputs)

    for h in hooks:
        h.remove()

    # Stack attention maps
    attn = torch.stack(attn_maps)   # [layers, batch, heads, tokens, tokens]

    # Fuse heads
    if head_fusion == "mean":
        attn = attn.mean(dim=2)
    elif head_fusion == "max":
        attn, _ = attn.max(dim=2)

    # Rollout
    eye = torch.eye(attn.size(-1)).to(attn.device)
    attn = attn + eye
    attn = attn / attn.sum(dim=-1, keepdim=True)

    joint_attn = attn[0]
    for i in range(1, attn.size(0)):
        joint_attn = attn[i].bmm(joint_attn)

    cls_attn = joint_attn[0, 0, 1:]
    num_patches = int(np.sqrt(cls_attn.shape[0]))

    mask = cls_attn.reshape(num_patches, num_patches).cpu().numpy()

    # Resize heatmap
    heatmap = cv2.resize(mask, img.size)
    heatmap = heatmap - heatmap.min()
    heatmap = heatmap / (heatmap.max() + 1e-8)

    heatmap_color = cv2.applyColorMap((heatmap * 255).astype(np.uint8), cv2.COLORMAP_JET)
    heatmap_color = cv2.cvtColor(heatmap_color, cv2.COLOR_BGR2RGB)

    blended = (0.5 * np.array(img) + 0.5 * heatmap_color).astype(np.uint8)

    return blended, heatmap

```

```python
##### evaluate test set

def evaluate_model(model, dataloader, num_labels_list, device):
    model.eval()
    correct_preds = [0] * len(num_labels_list)
    total_preds = [0] * len(num_labels_list)

    with torch.no_grad():
        for batch in dataloader:
            pixel_values = batch['pixel_values'].to(device)
            labels = [label.to(device) for label in batch['labels']]

            logits_list = model(pixel_values)

            for i, (logits, label) in enumerate(zip(logits_list, labels)):
                preds = torch.argmax(logits, dim=1)
                correct_preds[i] += (preds == label).sum().item()
                total_preds[i] += label.size(0)

    accuracies = [correct / total * 100 if total > 0 else 0.0
                  for correct, total in zip(correct_preds, total_preds)]
    return accuracies
```

```python
from sentence_transformers import SentenceTransformer, util

st_model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
def predict_answer(ques):
  max_similarity_attribute=""
  max_similarity_percentage=0

  similarity = util.cos_sim(st_model.encode("name of disease, name of skin condition", convert_to_tensor=True), st_model.encode(ques, convert_to_tensor=True)).item()
  if(similarity>max_similarity_percentage):
    max_similarity_percentage=similarity
    max_similarity_attribute="name"

  similarity = util.cos_sim(st_model.encode("cancerous, malignant, disease, skin condition", convert_to_tensor=True), st_model.encode(ques, convert_to_tensor=True)).item()
  if(similarity>max_similarity_percentage):
    max_similarity_percentage=similarity
    max_similarity_attribute="cancerous"

  similarity = util.cos_sim(st_model.encode("severity, seriousness, disease, skin condition", convert_to_tensor=True), st_model.encode(ques, convert_to_tensor=True)).item()
  if(similarity>max_similarity_percentage):
    max_similarity_percentage=similarity
    max_similarity_attribute="severity"

  similarity = util.cos_sim(st_model.encode("cause of disease/skin condition, reason of disease/skin condition", convert_to_tensor=True), st_model.encode(ques, convert_to_tensor=True)).item()
  if(similarity>max_similarity_percentage):
    max_similarity_percentage=similarity
    max_similarity_attribute="cause"

  similarity = util.cos_sim(st_model.encode("contagious, viral, disease, skin condition", convert_to_tensor=True), st_model.encode(ques, convert_to_tensor=True)).item()
  if(similarity>max_similarity_percentage):
    max_similarity_percentage=similarity
    max_similarity_attribute="contagious"

  similarity = util.cos_sim(st_model.encode("diagnosis, tests, exam, disease, skin condition", convert_to_tensor=True), st_model.encode(ques, convert_to_tensor=True)).item()
  if(similarity>max_similarity_percentage):
    max_similarity_percentage=similarity
    max_similarity_attribute="diagnosis"

  similarity = util.cos_sim(st_model.encode("prevention, disease, skin condition", convert_to_tensor=True), st_model.encode(ques, convert_to_tensor=True)).item()
  if(similarity>max_similarity_percentage):
    max_similarity_percentage=similarity
    max_similarity_attribute="prevention"

  return max_similarity_attribute

print(predict_answer("How serious is this disease")) # Checking function
```
